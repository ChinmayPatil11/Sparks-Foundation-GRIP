```python
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating the dataset
We use Scikit-Learn library to create the iris dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
df = pd.DataFrame(X,columns=iris.feature_names)
df['target'] = y
df['target'].replace(0,'setosa',inplace=True)
df['target'].replace(1,'versicolor',inplace=True)
df['target'].replace(2,'virginica',inplace=True)
df.head()
```

## Creating training set and test set
Using Scikit-Learn train_test_split library we create training and testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

## Training the Decision Tree Classifier
Using DecisionTreeClassifier we train the model on training set using default parameters

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

## Visualizing the Decision tree
We use the scikit-learn inbuilt plot_tree function from sklearn.tree to visualize the trained decision tree. Using this tree it becomes easy to understand how the classifier has trained

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(12,12))
plot_tree(clf,filled=True);
```

## Checking Accuracy of model
We check the accuracy of the model by using accuracy_score metric.

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
score = accuracy_score(y_test,y_pred)
print('Accuracy of the model:',score)
```
